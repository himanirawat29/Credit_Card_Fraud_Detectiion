```python
#import the libraries
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#create a dataframe using pandas
df = pd.read_csv('/kaggle/input/creditcardfraud/creditcard.csv')
df.head()
```

```python
#.shape will check the rows and columns 
df.shape
```

```python
#will check the null values
df.isnull().sum()
```

```python
df.info()
```

```python
#drop the time column
df.drop(['Time','Amount'] , axis = 1)
```

```python
#count the different values inside 'Class' column
df['Class'].value_counts()
```

```python
#will drop all duplicate values in the dataframe
df = df.drop_duplicates()
df.shape
```

```python
#No duplicate values as result shows False as an output
df.duplicated().any()
```

*****UnderSampling*****


```python
#will allocate all the values of 0 into actual and 1 into fraud
actual = df[df['Class'] == 0]
fraud = df[df['Class']== 1]
```

```python
print(actual.shape, fraud.shape)
```

```python
#countplot to visualizze the count into graph
sns.countplot(df , x = 'Class')

#this shows that target colummn is highly imbalanced
```

```python
#random sampling
Under_sample = actual.sample(n=473)
Under_sample.shape
```

```python
#add two data together by  using Concatenate function
new_df = pd.concat([Under_sample,fraud], axis = 0)
new_df['Class'].value_counts()
```

*** Training and Testing Data***

```python
#assign X and Y
X= new_df.drop('Class', axis = 1)
Y = new_df['Class']
```

```python
#split the dataset into 80% of training and 20% of testing set.
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size= 0.2, random_state = 2)
X.shape, X_test.shape, X_train.shape
```

```python
#use Logistic  Regression for Model Training 
log = LogisticRegression()
log.fit(X_train, Y_train)
```

```python
#Check the accuracy score fro predition dataset
Y_pred = log.predict(X_test)
accuracy_score(Y_test, Y_pred)
```

**Model Evaluation Using Precision, Recall, F1-Score**

```python
from sklearn.metrics import recall_score, precision_score, f1_score
```

```python
r1 = recall_score(Y_test, Y_pred)
print(f'recall score is {r1}')

p1 = precision_score(Y_test, Y_pred)
print(f'recall score is {p1}')

f1 = f1_score(Y_test, Y_pred)
print(f'recall score is {f1}')
```
